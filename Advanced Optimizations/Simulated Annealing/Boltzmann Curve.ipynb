```python
import numpy
import random
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    if x < -2 and x > 4:
        return 0
    else:
        return 4*np.sin(x**2) - (x*np.cos(x**2))
```

```python
fv = np.vectorize(f)
X = np.linspace(-2,4,1000)
y = []
i=1
```

```python
def b(x,T):
    if x>0:
        return 2.7**(-x/T)
    else:
        return 2.7**(x/T)
bV = np.vectorize(b)
```

```python
while i>0.0001:
    y.append(bV(fv(X),i))
    i = i*0.9
```

```python
y = np.array(y)
```

```python
plt.figure(figsize=(20,10))
plt.plot(X,fv(X),color="b")
plt.plot(X,y[0],color="g")
plt.plot(X,y[44],color="r")
plt.plot(X,y[-1],color="y")
```
