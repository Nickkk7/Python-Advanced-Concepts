```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
from collections import deque
```

```python
def f(x):
    return (1-np.cos(2*math.pi*0.01*x))*np.sin(2*math.pi*0.15*x)
```

```python
fv = np.vectorize(f)
```

```python
x = np.linspace(0,100,10000)
```

```python
plt.figure(figsize=(15,8))
plt.plot(x,fv(x))
```

```python
tabuList = deque()
T = 1
S = np.random.choice(x)
tabuList.append(S)
jumps = []
jumps.append(S)
```

```python
def neighbor_locator(S):
    global x
    index = 0
    for _ in x:
        if _ == S:
            break
        else:
            index = index+1
    return np.random.choice(x[index-500:index+500])
```

```python
while T > 0.0001:
    if len(tabuList) > 50:
        tabuList.popleft()
    R = neighbor_locator(S)
    for n in range(20):
        W = neighbor_locator(R)
        if W in tabuList and (fv(W) < fv(R) or R in tabuList):
            R = W
    if R not in tabuList:
        S = R
        tabuList.append(R)
        jumps.append(R)
    T = T*0.9
```
