```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    if x < -6 and x > 4:
        return 0
    else:
        return 4*np.sin(x**2) - (x*np.cos(x**2))
```

```python
fv = np.vectorize(f)
X = np.linspace(-2,4,1000)
```

```python
plt.plot(X,fv(X))
```

```python
def greedy_search(curve,start=-2,N=100):
    x = start
    trace = []
    for i in range(N):
        trace.append(x)
        u = 0.001
        xleft,xright = x-u,x+u
        yleft,yright = curve(xleft),curve(xright)
        if yleft > yright:
            x = xleft
        else:
            x = xright
    return x,trace
```

```python
xval,trace = greedy_search(fv,-2,15000)
```

```python
plt.plot(X,fv(X))
plt.scatter(xval,fv(xval),marker='X',s=100)
plt.plot(trace,fv(trace))
```
