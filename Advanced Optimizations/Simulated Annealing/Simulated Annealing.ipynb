```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
```

```python
def f(x):
    if x < -2 and x > 4:
        return 0
    else:
        return 4*np.sin(x**2) - (x*np.cos(x**2))
```

```python
fV = np.vectorize(f)
```

```python
def SA(search_space,func,T):
    scale = np.sqrt(T)
    start = np.random.choice(search_space)
    x = start * 1
    cur = func(x)
    history = [x]
    for i in range(6000):
        prop = x + np.random.normal()*scale
        if prop > 1 or prop < 0 or np.log(np.random.rand()) * T > (func(prop) - cur):
            prop = x
        x = prop
        cur = func(x)
        T  = 0.9*T
        history.append(x)
    return x,history
```

```python
X = np.linspace(-2,4,6000)
x1,history = SA(X,f,T=4)
```

```python
plt.plot(X,fV(X))
plt.scatter(x1,fV(x1),marker='x')
```
