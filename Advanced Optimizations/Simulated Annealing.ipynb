```python
import matplotlib.pyplot as plt
import numpy as np
import random
import math
```

```python
def f(x):
    return (1-np.cos(2*math.pi*0.01*x))*np.sin(2*math.pi*0.15*x)
```

```python
fv = np.vectorize(f)
```

```python
x = np.linspace(0,100,10000)
```

```python
plt.plot(x,fv(x))
```

```python
T = 1
S = np.random.choice(x)
best = fv(S)
jumps = []
```

```python
while T > 0.000001:
    R = np.random.choice(x)
    if fv(R) < fv(S) or random.random() < math.exp((fv(S) - fv(R))/T):
        S = R
        jumps.append(S)
    T = 0.9*T
```

```python
jumpsY = fv(jumps)
```

```python
plt.figure(figsize=(20,12))
plt.plot(x,fv(x))
plt.plot(jumps,jumpsY)
```
