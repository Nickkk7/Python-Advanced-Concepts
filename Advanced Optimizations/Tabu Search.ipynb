```python
import matplotlib.pyplot as plt
import numpy as np
import random
from collections import deque
```

```python
def f(x):
    if x > 10 or x < -10:
        return 0
    else:
        return x*np.cos(x) -x
fV = np.vectorize(f)
```

```python
x = np.linspace(-10,10,1000)
plt.figure(figsize=(20,10))
plt.plot(x,fV(x))
```

```python
tabu_location = deque()
bestX = np.random.randint(-10,10) 
bestY = f(bestX)
def tabu_Search(x):
    global tabu_location
    global bestX
    global bestY
    i = 0
    bestX = x
    bestY = f(bestX)
    while i<10000:
        potential_sol = np.linspace(bestX-2,bestX+2,20)
        potential_ans = fV(potential_sol)
        best_point = sorted(list(zip(potential_sol,potential_ans)),key = lambda x:x[1])
        tabu_location.append(best_point[0][0])
        for b in best_point:
            if b[0] in tabu_location or bestY > b[1]:
                pass
            else:
                bestX = b[0]
                bestY = b[1]
                break
        i = i+1
    print(bestX)
```

```python
tabu_Search(10)
```
